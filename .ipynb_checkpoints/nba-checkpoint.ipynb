```python
# Import Dependencies
import requests
import pandas as pd
```

```python
# Instantiate Function
def getInfo(position):
    
    # Base URL
    url = f'https://www.fantasypros.com/nba/stats/{position}.php'
    
    # Establish Header
    header = {
        "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
        "X-Requested-With": "XMLHttpRequest"
    }
    
    # Get Request
    r = requests.get(url, headers=header)
    
    # Read HTML
    df = pd.read_html(r.text)
    
    # Create DataFrame
    new_df = pd.DataFrame(df[0])
    
    # Return DataFrame
    return new_df
```

```python
# Get Overall Data
getInfo('overall')
```

```python
# Get Point Guard Data
getInfo('pg')
```
